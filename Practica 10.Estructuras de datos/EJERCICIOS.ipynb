```python
#Ejercicio 1:  conjunto de datos que contiene información de ventas mensuales.

class ErrorDeVenta(Exception):
    def __init__(self, message):
        super().__init__(message)
import pandas as pd
import random

meses = ["Ene", "Feb", "Mar", "Abr", "May", "Jun", "Jul", "Ago", "Sep", "Oct", "Nov", "Dic"]
ventas = [random.randint(1000, 3000) for i in range(12)]


df_ventas = pd.DataFrame({"Mes": meses, "Ventas": ventas})

def analisis_de_datos(df):
    try:
        mes_max_ventas = df[df["Ventas"] == df["Ventas"].max()]["Mes"].values[0]
        mes_min_ventas = df[df["Ventas"] == df["Ventas"].min()]["Mes"].values[0]
        return mes_max_ventas, mes_min_ventas
    except Exception:
        raise analisis_de_datos("No se pudieron analizar los datos de ventas.")

import matplotlib.pyplot as plt

def plot_datos(df):
    try:
        df.plot(kind='barh', x='Mes', y='Ventas')
        plt.xlabel('Ventas')
        plt.ylabel('Mes')
        plt.title('Ventas mensuales')
        plt.show()
    except Exception:
        raise ErrorDeVenta("No se pudo crear el gráfico de ventas.")

mes_max_ventas, mes_min_ventas = analisis_de_datos(df_ventas)
plot_datos(df_ventas)

```
